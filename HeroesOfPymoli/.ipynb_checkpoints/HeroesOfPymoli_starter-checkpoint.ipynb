```python
# Dependencies and Setup
import pandas as pd
import numpy as np

# File to Load (Remember to Change These)
file_to_load = "Resources/purchase_data.csv"

# Read Purchasing File and store into Pandas data frame
purchase_data = pd.read_csv(file_to_load)
purchase_data
#purchase_data["Price"].astype(float).map("${:.2f}".format)
```

## Player Count

* Display the total number of players


```python
purchase_data_player=purchase_data["SN"]
player_count=purchase_data_player.value_counts()
total_player_count=pd.DataFrame({"Total Players": [len(player_count)]})
total_player_count
```

## Purchasing Analysis (Total)

* Run basic calculations to obtain number of unique items, average price, etc.


* Create a summary data frame to hold the results


* Optional: give the displayed data cleaner formatting


* Display the summary data frame


```python
#get summary of Items
numberOfItems=purchase_data["Item Name"]
numberOfItems_unique=numberOfItems.value_counts()
CountOfItems_unique=len(numberOfItems_unique)
CountOfItems_unique
##get average & total price of items
avgPriceOfItem=purchase_data["Price"].mean()
TotalPriceOfItem=purchase_data["Price"].sum()
TotalPriceOfItem
#Get total number of purchases
NumOfPurchases=len(purchase_data["Purchase ID"].value_counts())
NumOfPurchases
Summary_data_frame=pd.DataFrame({"Number of Unique Items": [CountOfItems_unique], "Average Price": [avgPriceOfItem], "Number of Purchases":[NumOfPurchases], "Total Revenue": [TotalPriceOfItem]})
Summary_data_frame["Average Price"]=Summary_data_frame["Average Price"].map("${:.2f}".format)
Summary_data_frame["Total Revenue"]=Summary_data_frame["Total Revenue"].map("${:.2f}".format)
Summary_data_frame
```


## Purchasing Analysis (Gender)

* Run basic calculations to obtain purchase count, avg. purchase price, avg. purchase total per person etc. by gender




* Create a summary data frame to hold the results


* Optional: give the displayed data cleaner formatting


* Display the summary data frame

```python
#create purchase count by gender
all_gender=purchase_data.groupby('Gender')
player_count=all_gender['SN'].count()
player_count_df = pd.DataFrame(player_count)

# average price by gender
average_price = all_gender['Price'].mean()
average_price_df = pd.DataFrame(average_price)

# Total purchase price by all genders
total_gender_price = all_gender['Price'].sum()

#  average price per person 
avg_price_person =(all_gender["Price"].sum()/all_gender["SN"].count())
average_price_person = total_gender_price / player_count

#Adding average price to table
player_count_df['Average Purchase Price'] = average_price.map("${:.2f}".format)
gender_purchase_df = pd.DataFrame(player_count_df)
gender_purchase_df

gender_purchase_df_renamed=gender_purchase_df.rename(columns={'SN':'Purchase Count'})
gender_purchase_df_renamed

# Add 'Total Purchase Price' to table
gender_purchase_df_renamed['Total Purchase Price'] = total_gender_price.map("${:.2f}".format)
gender_purchase_df_1 = pd.DataFrame(gender_purchase_df_renamed)

# Add 'Average Price person' to table
gender_purchase_df_1['Average Price Per Person'] = average_price_person.map("${:.2f}".format)
total_gender_purchase_df = pd.DataFrame(gender_purchase_df_1)
total_gender_purchase_df
```

## Age Demographics

* Establish bins for ages


* Categorize the existing players using the age bins. Hint: use pd.cut()


* Calculate the numbers and percentages by age group


* Create a summary data frame to hold the results


* Optional: round the percentage column to two decimal points


* Display Age Demographics Table


```python
#Bin for age
bin_for_age= [0, 9.90, 14.90, 19.90, 24.90, 29.90, 34.90, 39.90, 99999]
group_names = ["<10", "10-14", "15-19", "20-24", "25-29", "30-34", "35-39", "40+"]
player_count
# Define Age group
purchase_data['Total Count'] = pd.cut(purchase_data['Age'],
                                     bin_for_age, labels=group_names)
age_group=purchase_data.groupby('Total Count')
age_group_df=age_group.count()

age_df=age_group_df[['Age']]
age_df
```

* Create a summary data frame to hold the results


* Optional: give the displayed data cleaner formatting

```python
# Bins for different ages
age_bins2 = [0, 10, 15, 20, 25, 30, 35, 40, 100]
group_names2 = ['<10', '10-14', '15-19', '20-24', '25-29', '30-34', '35-39', '40+']

# Age groups
purchase_data['Total Count'] = pd.cut(purchase_data['Item ID'],
                                     age_bins2, labels=group_names2)
age_groups2 = purchase_data.groupby('Total Count')
age_groups2_df = age_groups2.count()
age_purchase = age_groups2_df['Item ID']
age_purchase

age2_df = age_groups2_df[['Item ID']]
age2_df
```

## Top Spenders

* Run basic calculations to obtain the results in the table below


* Create a summary data frame to hold the results


* Sort the total purchase value column in descending order


* Optional: give the displayed data cleaner formatting


* Display a preview of the summary data frame



```python
# Total Purchase Count
count_per_SN = purchase_data.groupby('SN')['Price'].count()
count_per_SN_df = pd.DataFrame(count_per_SN)

# Total Purchase Price
unique_SN_df = purchase_data.groupby('SN')['Price'].sum() #.map("${:.2f}".format)

# Average Purchase Price
avg_price_per_SN = purchase_data.groupby('SN')['Price'].mean() #.map("${:.2f}".format)

# Add 'Average Purchase Price'
count_per_SN_df['Average Purchase Price'] = avg_price_per_SN #.map("${:.2f}".format)

# Add 'Total Purchase Value'
count_per_SN_df['Total Purchase Value'] = unique_SN_df #.map("${:.2f}".format)
count_per_SN_df

# Rename 'Price' to 'Purchase Count'
updated_count_per_SN_df = count_per_SN_df.rename(columns={
    'Price': 'Purchase Count',
})

# Sort 'Price' from highest to lowest
sorted_SN_df = updated_count_per_SN_df.sort_values(by=['Total Purchase Value'], ascending=False)
sorted_SN_df

# Show top 5 spenders only
sorted_SN_df.head(5)

# The top 5 spenders bought less than $5 per item with less than $20 purchase value.
```

## Most Popular Items

* Retrieve the Item ID, Item Name, and Item Price columns


* Group by Item ID and Item Name. Perform calculations to obtain purchase count, average item price, and total purchase value


* Create a summary data frame to hold the results


* Sort the purchase count column in descending order


* Optional: give the displayed data cleaner formatting


* Display a preview of the summary data frame



```python
# Total Purchase Count
count_per_item = purchase_data.groupby('Item ID')['Price'].count()
count_per_item_df = pd.DataFrame(count_per_item)

# Total Purchase Price
unique_item_df = purchase_data.groupby('Item ID')['Price'].sum()

# Add Item Name
item_name_df = purchase_data.iloc[:, 5]
count_per_item_df['Item Name'] = item_name_df

# Add Item Price
item_price_df = purchase_data.iloc[:, 6]
count_per_item_df['Item Price'] = item_price_df

# Add Total Purchase Value
count_per_item_df['Total Purchase Value'] = unique_item_df
count_per_item_df

# Rename 'Price' to 'Purchase Count'
updated_count_per_item_df = count_per_item_df.rename(columns={
    'Price': 'Purchase Count',
})

# Rearrange columns
organized_items_df = updated_count_per_item_df[['Item Name', 'Purchase Count', 'Item Price', 'Total Purchase Value']]

# Sort 'Purchase Count' from highest to lowest
sorted_items_df = organized_items_df.sort_values(by=['Purchase Count'], ascending=False)

# Show top 5 popular items only
sorted_items_df.head()
```

## Most Profitable Items
